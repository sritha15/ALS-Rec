```python
import pandas as pd
train_set=pd.read_csv("product_info")
```

```python
train_set.columns
```

```python
df = train_set.loc[train_set['secondary_category'] == 'Moisturizers', ['secondary_category', 'product_id']]
```

```python
df.isnull().sum()
```

```python
print(df['secondary_category'].unique())
```

```python
print(df.describe())
```

```python
r1=pd.read_csv("reviews_0-250.csv")
r2=pd.read_csv("reviews_250-500.csv")
r3=pd.read_csv("reviews_500-750.csv")
r4=pd.read_csv("reviews_750-1250.csv")
r5=pd.read_csv("reviews_1250-end.csv")
```

```python
r3.head()
```

```python
newr1 = pd.merge(r1, df[['product_id']], on='product_id', how='inner')
newr2 = pd.merge(r2, df[['product_id']], on='product_id', how='inner')
newr3 = pd.merge(r3, df[['product_id']], on='product_id', how='inner')
newr4 = pd.merge(r4, df[['product_id']], on='product_id', how='inner')
newr5 = pd.merge(r5, df[['product_id']], on='product_id', how='inner')

```

```python
newr5
```

```python
newr1.to_csv('r1.csv', index=False)
newr2.to_csv('r2.csv', index=False)
newr3.to_csv('r3.csv', index=False)
newr4.to_csv('r4.csv', index=False)
newr5.to_csv('r5.csv', index=False)
```
